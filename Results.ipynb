# Results for comparative implementation of Dijkstra's algorithm using Priority queue, Networkx, and Neo4j APOC
### The runtime observations were and noted and drafted in a separate csv `results.csv` which is being used here to visualize the results

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import and read the csv

results = pd.read_csv('results.csv')   # change it as per your file path
results
```

RUNTIME COMPARISON WITH PRIORITY QUEUE (SCRATCH) AND NETWORKX LIBRARY FOR GRAPHS WITH 1000 AND 10,000 NODES

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Plot for num_nodes = 1000
axes[0].plot(results['size_1000'], results['time_scratch_1000'], 'o-', label = 'Priority queue', color = 'orange')
axes[0].plot(results['size_1000'], results['time_nx_1000'], 'o-', label = 'Networkx', color = 'blue')
axes[0].text(3.5, 0.117, 'num_nodes = 1000', fontsize = 11.5, color = 'black', weight = 'bold')
axes[0].set_xlabel('Number of edges (tens of thousand)', fontsize = 14)
axes[0].set_ylabel('Runtime (s)', fontsize=14)
axes[0].set_ylim(-0.005, 0.13)
axes[0].tick_params(axis='both', labelsize=12)
axes[0].grid()
axes[0].legend()

# Plot for num_nodes = 10,000
axes[1].plot(results['size_10k'], results['time_scratch_10k'], 'o-', label = 'Priority queue', color = 'orange')
axes[1].plot(results['size_10k'], results['time_nx_10k'], 'o-', label = 'Networkx', color = 'blue')
axes[1].text(35, 15.5, 'num_nodes = 10,000', fontsize = 11.5, color = 'black', weight = 'bold')
axes[1].set_xlabel('Number of edges (tens of thousand)', fontsize = 14)
axes[1].set_ylabel('Runtime (s)', fontsize=14)
axes[1].set_ylim(-1, 17)
axes[1].tick_params(axis='both', labelsize=12)
axes[1].grid()
axes[1].legend()

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

RUNTIME COMPARISON WITH NETWORKX LIBRARY AND NEO4J APOC LIBRARY FOR GRAPHS WITH 1000 AND 10,000 NODES

```python
fig, axes = plt.subplots(1, 2, figsize = (14, 4))

# Plot for num_nodes = 1000
ax1 = axes[0]
ax1.plot(results['size_1000'], results['time_nx_1000'], 'o-', label='Networkx', color = 'blue')
ax1.text(3.2, 0.065, 'num_nodes = 1000', fontsize = 12, color = 'black', weight = 'bold')
ax1.set_xlabel('Number of edges (tens of thousand)', fontsize = 14)
ax1.set_ylabel('Networkx Runtime (s)', fontsize = 14, color = 'blue', labelpad = 10)
ax1.tick_params(axis = 'both', labelsize = 12)
ax1.set_ylim(-0.005, 0.075)
ax1.grid()

# Create a twin Axes for the second plot
ax2 = ax1.twinx()
ax2.plot(results['size_1000'], results['time_apoc_1000'], 'o-', label = 'APOC', color = 'red')
ax2.set_ylabel('APOC Runtime (s)', fontsize = 14, color = 'red', labelpad = 10)
ax2.tick_params(axis = 'both', labelsize = 12)
ax2.set_ylim(-0.00025, 0.0037)

# Plot for num_nodes = 10,000
ax1 = axes[1]
ax1.plot(results['size_10k'], results['time_nx_10k'], 'o-', label = 'Networkx', color = 'blue')
ax1.text(32, 0.65, 'num_nodes = 10,000', fontsize = 12, color = 'black', weight = 'bold')
ax1.set_xlabel('Number of edges (tens of thousand)', fontsize = 14)
ax1.set_ylabel('Networkx Runtime (s)', fontsize = 14, color = 'blue', labelpad = 10)
ax1.tick_params(axis = 'both', labelsize = 12)
ax1.set_ylim(-0.05, 0.75)
ax1.grid()

# Create a twin Axes for the second plot
ax2 = ax1.twinx()
ax2.plot(results['size_10k'], results['time_apoc_10k'], 'o-', label = 'APOC', color = 'red')
ax2.set_ylabel('APOC Runtime (s)', fontsize = 14, color = 'red', labelpad = 10)
ax2.tick_params(axis = 'both', labelsize = 12)
ax2.set_ylim(0.0015, 0.00265)

# Adjust layout
plt.tight_layout()

plt.subplots_adjust(wspace = 0.425)

# Show the plots
plt.show()
```
